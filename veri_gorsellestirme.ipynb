```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
veri = pd.read_csv("kalp_rahatsizligi.csv",sep=",")
veri.head(3)
```

```python
veri.describe()
```

```python
veri.info()
```

```python
sns.histplot(data=veri,x=veri.yas,bins=10)
plt.xlabel("Yaş")
plt.ylabel("frekans/adet")
plt.show()
```

```python
# serum_kolestrol verisinin histogram grafiğini çizdiriniz.
sns.histplot(data=veri,x=veri.serum_kolestrol,bins=20)
plt.ylabel("frekans/adet")
plt.xlabel("Serum kolestrol")
plt.show()
```

```python
# serum_kolestrol verisinin histogram grafiğini cisniyete göre  çizdiriniz.

sns.histplot(data=veri,x=veri.serum_kolestrol,bins=20,hue=veri.gogus_agrisi_tipi,element="step",palette="bright")
plt.ylabel("frekans/adet")
plt.xlabel("Serum kolestrol")
plt.show()

```

```python
sns.countplot(data=veri,x=veri.cinsiyet)
plt.xlabel("cinsiyet")
plt.ylabel("adet")
plt.show()
```

```python
sns.countplot(data=veri,x=veri.kalp_rahatsizligi)
plt.xlabel("kalp rahatsızlığı")
plt.ylabel("adet")
plt.show()
```

```python
sns.pairplot(data=veri)
plt.show()
```

```python
sns.scatterplot(data=veri,x=veri.en_yuksek_kalp_hizi,y=veri.yas)
plt.show()
```

```python
sns.lmplot(data=veri,x="yas",y="en_yuksek_kalp_hizi",hue="cinsiyet")
plt.show()
```
